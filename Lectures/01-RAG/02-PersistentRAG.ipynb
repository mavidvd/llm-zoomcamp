# 2. Persistent RAG

This RAG uses the library sqlitesearch which stores a database in a .db file, which allows data to keep the knowledge base across sessions.

```python
from dotenv import load_dotenv
load_dotenv()

from _ingest import load_db_knowledge_base
from _rag_helper import RAGBase
from openai import OpenAI

sqlite_index = load_db_knowledge_base(course = 'machine-learning-zoomcamp')
```

```python
openai_client = OpenAI()

assistant = RAGBase(
    index=sqlite_index,
    llm_client=openai_client,
    input = 'qa'
)

answer = assistant.rag("Are we going to use Jupytr Notebooks?")
print(answer)
```
